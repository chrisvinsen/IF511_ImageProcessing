```python
from PIL import Image
```

```python
img = Image.open("mario.jpg")
```

```python
new_size = (1200,1200)
# img.crop receive (l,t,r,b)
img2 = img.crop((75,40,200,175))
img2 = img2.resize(new_size, resample= Image.NEAREST)

img3 = img.crop((75,40,200,175))
img3 = img3.resize(new_size, resample= Image.BILINEAR)
```

```python
fig, ax = plt.subplots(1,2,figsize=(18,18)) 

ax[0].imshow(img2) 
ax[0].axis('off') 

ax[1].imshow(img3)
ax[1].axis('off')

plt.show() 
```

```python
new_size = (1200,1200)
# img.crop receive (l,t,r,b)
img2 = img.crop((75,40,200,175))
img2 = img2.resize(new_size, resample= Image.BILINEAR)

img3 = img.crop((75,40,200,175))
img3 = img3.resize(new_size, resample= Image.BICUBIC)
```

```python
import numpy as np

img = np.array([
    [[0,255,255],[255,255,0]],
    [[255,0,255],[0,0,0]],
])
```

```python
img.shape
```

```python
img = Image.fromarray(np.uint8(img)).convert('RGB')
new_size = (5,5)

img_nearest = img.resize(new_size, resample= Image.NEAREST)
img_bilinear = img.resize(new_size, resample= Image.BILINEAR)
img_bicubic = img.resize(new_size, resample= Image.BICUBIC)
```

```python
fig, ax = plt.subplots(2,2,figsize=(18,18)) 

ax[0][0].imshow(img) 
ax[0][0].axis('off') 

ax[0][1].imshow(img_nearest)
ax[0][1].axis('off')

ax[1][0].imshow(img_bilinear)
ax[1][0].axis('off')

ax[1][1].imshow(img_bicubic)
ax[1][1].axis('off')

plt.show() 
```
