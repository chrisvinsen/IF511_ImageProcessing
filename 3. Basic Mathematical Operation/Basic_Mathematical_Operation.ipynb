# Pixel Addition

```python
from PIL import Image, ImageOps
gray_img = Image.open('smile1.png')
```

```python
grayscale = ImageOps.grayscale(gray_img)
grayscale = np.array(grayscale)
grayscale
```

```python
fig, ax = plt.subplots(figsize=(5,5)) 

ax.imshow(grayscale, cmap='gray') 

plt.title('gray_img', fontweight ="bold",fontsize=15) 
plt.show() 
```

```python
grayscale[200][100]
```

```python
result = np.add(grayscale, 100)

fig, ax = plt.subplots(figsize=(5,5)) 

ax.imshow(result, cmap='gray', vmin=0, vmax=255) 
ax.axis('off') 

plt.title('gray_img + 100', fontweight ="bold",fontsize=15) 
plt.show() 
```

```python
result[200][100]
```

```python
smile1=plt.imread('smile1.png')
smile2=plt.imread('smile2.png')

fig, ax = plt.subplots(1,2,figsize=(5,5)) 

ax[0].imshow(smile1, cmap='gray') 
ax[0].axis('off') 
ax[0].set_title('smile 1', fontsize = 15)

ax[1].imshow(smile2, cmap='gray') 
ax[1].axis('off') 
ax[1].set_title('smile 2', fontsize = 15)

plt.show() 
```

```python
result = smile1 + smile2

fig, ax = plt.subplots(figsize=(5,5)) 

ax.imshow(result, cmap='gray') 
ax.axis('off') 

plt.title('smile 1 + smile 2', fontweight ="bold",fontsize=15) 
plt.show() 
```

# Image Subtraction

```python
result[200][100]
```

```python
result = smile1 - smile2

fig, ax = plt.subplots(figsize=(5,5)) 

ax.imshow(result, cmap='gray') 
ax.axis('off') 

plt.title('smile 1 - smile 2', fontweight ="bold",fontsize=15) 
plt.show() 
```

# Image Division

```python
result = np.divide(grayscale, 2)

fig, ax = plt.subplots(figsize=(5,5)) 

ax.imshow(result, cmap='gray', vmin=0, vmax=255) 
ax.axis('off') 

plt.title('gray_img / 2', fontweight ="bold",fontsize=15) 
plt.show() 
```

```python
result[200][100]
```

```python
result = np.divide(smile1, smile2, where=smile2!=0)

fig, ax = plt.subplots(figsize=(5,5)) 

ax.imshow(result, cmap='gray') 
ax.axis('off') 

plt.title('smile 1 / smile 2', fontweight ="bold",fontsize=15) 
plt.show() 
```

# Image Translation

```python
import cv2 
import numpy as np 
  
img = cv2.imread('daisy.png')
```

```python
height, width = img.shape[:2] 
T = np.float32([[1, 0, 20], [0, 1, 50]])
result = cv2.warpAffine(img, T, (width, height))
```

```python
result
```

```python
fig, ax = plt.subplots(figsize=(5,5)) 

ax.imshow(result, cmap='gray') 
ax.axis('off') 

plt.title('translation', fontweight ="bold",fontsize=15) 
plt.show() 
```

# Image Rotation

```python
result = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)

fig, ax = plt.subplots(figsize=(5,5)) 

ax.imshow(result, cmap='gray') 
ax.axis('off') 

plt.title('rotation', fontweight ="bold",fontsize=15) 
plt.show() 
```

# Image Scaling

```python
result = cv2.resize(img, (780, 540), interpolation = cv2.INTER_NEAREST) 
```

```python
fig, ax = plt.subplots(figsize=(5,5)) 

ax.imshow(result, cmap='gray') 

plt.title('Scaling', fontweight ="bold",fontsize=15) 
plt.show() 
```

# BITWISE OPERATION

```python
lines = cv2.imread('lines.png')
```

```python
result = cv2.bitwise_or(img, lines)
```

```python
fig, ax = plt.subplots(figsize=(5,5)) 

ax.imshow(result, cmap='gray') 

plt.title('BITWISE OR', fontweight ="bold",fontsize=15) 
plt.show()
```

```python
result = cv2.bitwise_and(img, lines)
```

```python
fig, ax = plt.subplots(figsize=(5,5)) 

ax.imshow(result, cmap='gray') 

plt.title('BITWISE AND', fontweight ="bold",fontsize=15) 
plt.show()
```

```python
result = cv2.bitwise_not(img)
```

```python
fig, ax = plt.subplots(figsize=(5,5)) 

ax.imshow(result, cmap='gray') 

plt.title('INVERSE', fontweight ="bold",fontsize=15) 
plt.show()
```
